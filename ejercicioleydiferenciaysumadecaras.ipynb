*   LA DIFERENCIA DE LAS CARAS SEA MAYOR QUE 1



```python
from itertools import product
from fractions import Fraction
```

```python
def P(A, Omega):
   P= Fraction(len(A), len(Omega) )
   return P
```

```python
S_1 = {(i,j) for i in range(1,7) for j in range (1,7) if abs(i-j)>1}
S_1
```

```python
L = [i for i in range(1,7)]
Omega = set(product(L, repeat =2))
Omega
```

```python
M = {(i,j): i-j for i,j in Omega if abs(i-j)>1 }
M
```

```python
from collections import defaultdict
dM = defaultdict(set)
dM
```

```python
M.items()
```

```python
for l, v in M.items():
 dM[v]. add(l)
```

```python
dM
```

```python
dM.items()
```

```python
leyM = {k : P(A,Omega) for k, A in dM.items()}
leyM
```

```python
import pandas as pd
```

```python
leyM = pd.Series(leyM)
leyM = leyM.sort_index()
```

```python
leyM
```

```python
sumaM = leyM.sum()
sumaM
```

```python
N= {(i,j): i-j for i, j in Omega if abs(i-j)<2}
N
```

```python
from collections import defaultdict
dN = defaultdict(set)
dN
```

```python
N.items()
```

```python
for l,v in N.items():
   dN[v].add(l)
```

```python
dN
```

```python
dN.items()
```

```python
leyN = {k : P(A,Omega) for k , A in dN.items() }
leyN
```

```python
leyN = pd.Series(leyN)
leyN = leyN.sort_index()
```

```python
leyN
```

```python
sumaN= leyN.sum()
sumaN
```

```python
int(sumaM+sumaN)
```

**Y con esto se verifica que la sumatoria de las probabilidades es igual a 1, sumando la probabilidad de que ocurra que la diferencia de las dos caras sea mayor a uno y la probabilidad de que esto no ocurra.**
